```python
from numpy import savez_compressed
#training......
#1.loading the data( our own data)
import pandas as pd
file=pd.read_csv("/content/irisexcel(1).csv")
x=file[["sepal_length","sepal_width","petal_length","petal_width"]]
y=file["species"]
#cross validation
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)

#2.load the model(preceptron)
from sklearn.linear_model import Perceptron
model=Perceptron()
#3.fit data into model
model.fit(xtrain,ytrain)
#validation
op=model.predict(xtest)
from sklearn.metrics import accuracy_score
score=accuracy_score(ytest,op)
print("Accuracy score=",score*100)
#testing...predict
op2=model.predict([[3.5,4.5,5.5,6.7]])
print(op2)
```

```python
from numpy import savez_compressed
#training......
#1.loading the data( our own data)
import pandas as pd
file=pd.read_csv("/content/13 MLP winequality-red.csv")
x=file[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides"]]
y=file["quality"]
#cross validation
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)

#2.load the model(preceptron)
from sklearn.linear_model import Perceptron
model=Perceptron()
#3.fit data into model
model.fit(xtrain,ytrain)
#validation
op=model.predict(xtest)
from sklearn.metrics import accuracy_score
score=accuracy_score(ytest,op)
print("Accuracy score=",score*100)
#testing...predict
op2=model.predict([[3.5,4.5,5.5,6.7,3.6]])
print(op2)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm=confusion_matrix(ytest,op)
print(cm)
cmd= ConfusionMatrixDisplay(confusion_matrix=cm,

                              display_labels=model.classes_)
cmd.plot()
```
