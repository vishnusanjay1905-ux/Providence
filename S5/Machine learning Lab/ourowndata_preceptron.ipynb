```python
#training......
#1.loading the data( our own data)
x=[[1,1],[15,16],[1,2],[18,19],[5,6]]
y=["boy","girl","boy","girl","boy"]
#2.load the model(preceptron)
from sklearn.linear_model import Perceptron
model=Perceptron()
#3.fit data into model
model.fit(x,y)
#testing...predict
op=model.predict([[3,4]])
print(op)
```
